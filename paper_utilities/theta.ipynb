```python
import mpmath as mp
import numpy as np
import matplotlib.pyplot as plt

# Ustawiamy precyzję obliczeń
mp.mp.dps = 200 

def theta(k, lam, s_min=-100, s_max=100):
    total = mp.mpf('0')
    for i in range(s_min, s_max + 1):
        total += lam * mp.power(k, i) * mp.exp(-lam * mp.power(k, i))
    return total

k = mp.mpf('2')

# Pierwszy zakres λ
# lambdas1 = [mp.mpf(x) for x in np.linspace(1, 5, 500)]
# thetas1 = [theta(k, lam) for lam in lambdas1]

# Drugi zakres λ
lambdas2 = [mp.mpf(x) for x in np.linspace(500, 1500, 500)]
thetas2 = [theta(k, lam) for lam in lambdas2]
```

```python

# Konwersja do floatów dla matplotlib
l1 = [float(l) for l in lambdas1]
t1 = [float(t) for t in thetas1]
l2 = [float(l) for l in lambdas2]
t2 = [float(t) for t in thetas2]

# Wykres 1
plt.figure(figsize=(6, 4))
plt.plot(l1, t1)
plt.title(r'$\Theta(2, \lambda)$ dla $\lambda \in [0.01, 5]$')
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$\Theta(2, \lambda)$')
plt.tight_layout()
plt.grid()
plt.show()

# Wykres 2
plt.figure(figsize=(6, 4))
plt.plot(l2, t2)
plt.title(r'$\Theta(2, \lambda)$ dla $\lambda \in [5, 100]$')
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$\Theta(2, \lambda)$')
plt.tight_layout()
plt.grid()
plt.show()
```
